# График бенчмарка HW8

Строит график поиска подстроки по `data/hw8_fm_index_benchmark.csv`.

```python
from csv import DictReader
from pathlib import Path

import matplotlib.pyplot as plt

csv_path = Path("data/hw8_fm_index_benchmark.csv")
rows = []
with csv_path.open(newline="", encoding="utf-8") as file:
    reader = DictReader(file)
    for row in reader:
        rows.append(row)

xs = [int(row["pattern_length"]) for row in rows]
ys = [float(row["query_ms_per_query"]) for row in rows]

plt.figure(figsize=(8, 5))
plt.plot(xs, ys, marker="o", linewidth=2)
plt.title("Поиск подстроки в FM-index")
plt.xlabel("Длина шаблона")
plt.ylabel("Среднее время запроса, мс")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("data/hw8_fm_index_benchmark.png", dpi=160)
plt.show()
```
