# График бенчмарка HW7

Строит графики поиска ближайшего соседа по `data/hw7_kdtree_benchmark.csv`.

```python
from csv import DictReader
from pathlib import Path

import matplotlib.pyplot as plt

csv_path = Path("data/hw7_kdtree_benchmark.csv")
rows = []
with csv_path.open(newline="", encoding="utf-8") as file:
    reader = DictReader(file)
    for row in reader:
        rows.append(row)

xs = [int(row["dimensions"]) for row in rows]
ys_time = [float(row["query_ms_per_query"]) for row in rows]
ys_reads = [float(row["avg_node_reads_per_query"]) for row in rows]

fig, (ax_time, ax_reads) = plt.subplots(2, 1, figsize=(8, 8), sharex=True)
ax_time.plot(xs, ys_time, marker="o", linewidth=2)
ax_time.set_title("Поиск ближайшего соседа в k-d tree")
ax_time.set_ylabel("Среднее время запроса, мс")
ax_time.grid(True, alpha=0.3)

ax_reads.plot(xs, ys_reads, marker="o", linewidth=2, color="tab:orange")
ax_reads.set_xlabel("Размерность k")
ax_reads.set_ylabel("Чтений узлов за запрос")
ax_reads.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("data/hw7_kdtree_benchmark.png", dpi=160)
plt.show()
```
